#### Merges the OECD's life satisfaction data and the IMF's GDP per capita data

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    '''The oecd_bli data frame is being filtered to only include rows where 
        the value in the "INEQUALITY" column is equal to "TOT"'''
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    
    ''' - The index argument is set to "Country", meaning that the unique values in the "Country" 
          column will become the new index of the pivoted data frame.
        - The columns argument is set to "Indicator", meaning that the unique values in the 
          "Indicator" column will become the new columns of the pivoted data frame.
        - Finally, the values argument is set to "Value", meaning that the values in the "Value" 
          column will become the values in the pivoted data frame.'''
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    
    '''
    The columns argument is a dictionary that maps the original column name ("2015") to the 
    new column name ("GDP per capita")
    '''
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    
    '''
    making the "Country" column the new index of the gdp_per_capita DataFrame, effectively 
    replacing the default integer-based index with the values in the "Country" column.
    '''
    gdp_per_capita.set_index("Country", inplace=True)
    
    ''' The left_index and right_index arguments are both set to True, meaning that the indexes 
        of the two DataFrames are used as the join keys. 
        The result includes only the rows for which the index values exist in both DataFrames.'''
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    
    # Sort the data frame by 'GDP per capita'
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    
    # subtract the set of remove_indices from the set of all 36 indices.
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    
    '''Select the columns ("GDP per capita", 'Life satisfaction'), and list out 
        the rows from 'keep_indices' 
        - Life satisfaction is one of the columns in ' columns="Indicator" ' 
    '''
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

#### Overfitting the Model

- This code is likely to **overfit the model** because it is **using a high-degree polynomial (degree=30) as a feature transformer in the pipeline**. 
    - **Overfitting occurs** when a model is **too complex and fits the training data too well**, capturing not just the underlying trend but also the noise in the data.
<br></br>
- In this case, the **polynomial features** with a **degree of 30** will **generate a large number of additional features from the input data**, making the **model extremely complex**. 
    - When the model is trained on the training data, it is likely to **fit the training data very well**, but **not generalize well to new dat**a, resulting in **high variance and overfitting**.
<br></br>
- To **avoid overfitting**, it is recommended to **use simpler models** with **fewer features** or to use **regularization techniques** to **penalize the complexity of the model**. 
    - In this case, **reducing the degree of the polynomial** or using a **simpler feature transformation method** such as a **logarithmic transformation** may help **reduce overfitting**.

```python
full_country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(8,3))
plt.axis([0, 110000, 0, 10])

from sklearn import preprocessing
from sklearn import pipeline

'''
Generate polynomial features from the input data
- The 'degree' parameter specifies the degree of the polynomial to be generated (e.g. degree=30 means that features up to the 30th power will be generated). 
- The 'include_bias' parameter is set to False, meaning that a bias term (also known as an intercept term) will not be included in the generated polynomial features.
'''
poly = preprocessing.PolynomialFeatures(degree=30, include_bias=False)

'''
Scales the input data by subtracting the mean and dividing by the standard deviation, so that the transformed data has zero mean and unit variance.
- Ensure that the input features are in a standard scale and to prevent large input values from dominating the model training process.
'''
scaler = preprocessing.StandardScaler()

'''
Fits a linear equation to the input data and predicts the target values based on the learned coefficients.
'''
lin_reg2 = linear_model.LinearRegression()

'''
The pipeline will apply these components in the order they are specified, so the 
    1. input data will first be transformed by the PolynomialFeatures preprocessor, 
    2. then by the StandardScaler preprocessor, and 
    3. finally passed to the LinearRegression model for prediction.
'''
pipeline_reg = pipeline.Pipeline([('poly', poly), ('scal', scaler), ('lin', lin_reg2)])
pipeline_reg.fit(Xfull, yfull)

'''
The input data X is transformed using the np.newaxis method from the NumPy library. 
    - The np.newaxis method is used to increase the dimensionality of an array. 
        - In this case, it is used to convert the 1-dimensional array X into a 2-dimensional array 
            with a single column. 
            - This is necessary because the "predict method of the pipeline" 'expects a 2-dimensional array' as input, 
              with the number of rows representing the number of samples and 
                  the number of columns representing the number of features.
'''
curve = pipeline_reg.predict(X[:, np.newaxis])
plt.plot(X, curve)
plt.xlabel("GDP per capita (USD)")
save_fig('overfitting_model_plot')
plt.show()
```

#### Reduce Overfitting the Model

```python
# Extract the relevant columns as arrays
Xfull = np.c_[full_country_stats["GDP per capita"]]
yfull = np.c_[full_country_stats["Life satisfaction"]]

# Split the data into training and test sets
n_samples = len(full_country_stats)
X_train = Xfull[:-n_samples // 5]
y_train = yfull[:-n_samples // 5]
X_test = Xfull[-n_samples // 5:]
y_test = yfull[-n_samples // 5:]

# Use a logarithmic transformation to reduce the impact of outliers
'''
One thing to keep in mind is that you should make sure that the input features are positive and 
non-zero before applying the logarithmic transformation. If you have any zero or negative values, 
you may want to add a small constant to the input values before applying the transformation to ensure 
that the transformed values are always positive.
'''
X_train_log = np.log(X_train)

# Fit a linear regression model
lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_train_log, y_train)

# Predict the target values based on the learned coefficients
X_test_log = np.log(X_test)
y_pred = lin_reg.predict(X_test_log)

# Plot the results
plt.scatter(X_test_log, y_test, color='black')
plt.plot(X_test_log, y_pred, color='blue', linewidth=3)
plt.xlabel("GDP per capita (log scale)")
plt.ylabel("Life satisfaction")
plt.show()
```
